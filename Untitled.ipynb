```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('http?://\S+|www\.\S+','',text)
    text = re.sub('<.*?>+','', text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'', text)
    text = re.sub('\n','', text)
    text = re.sub('\w*\d\w*','', text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```
